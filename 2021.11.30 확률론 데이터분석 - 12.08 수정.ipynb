## 정규화

```python
transformer = MinMaxScaler()
using_time = np.array(picher['사용시간']).reshape(-1, 1)
transformer.fit(using_time) 
picher['사용시간'] = transformer.transform(using_time)
```

```python
transformer = MinMaxScaler()
pc = np.array(picher['롤 점유율']).reshape(-1, 1)
transformer.fit(pc)
picher['롤 점유율'] = transformer.transform(pc)
```

```python
transformer = MinMaxScaler()
per = np.array(picher['롤 전주대비 증감률']).reshape(-1, 1)
transformer.fit(per)
picher['롤 전주대비 증감률'] = transformer.transform(per)
```

```python
transformer = MinMaxScaler()
per = np.array(picher['코로나']).reshape(-1, 1)
transformer.fit(per)
picher['코로나'] = transformer.transform(per)
```

```python
transformer = MinMaxScaler()
per = np.array(picher['증감률']).reshape(-1, 1)
transformer.fit(per)
picher['증감률'] = transformer.transform(per)
```

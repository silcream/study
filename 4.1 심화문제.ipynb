```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math
```

```python
def mySin(x):      # sin(x)
  return np.sin(x)

def myCos(x):      # cos(x)
  return np.cos(x)

x = np.linspace(-np.pi, np.pi)
y_sin = mySin(x)
y_cos = myCos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
import plotly.graph_objects as go
  
feature_x = np.arange(0, np.pi) 
feature_y = np.arange(0, 20) 
  
# Creating 2-D grid of features 
[X, Y] = np.meshgrid(feature_x, feature_y) 
  
Z = np.cos(X / 10) + np.sin(Y / 10) 
  
fig = go.Figure(data =
     go.Heatmap(x = feature_x, y = feature_y, z = Z,))
  
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as pp
 
def func(x, y):
    f = x ** 2 + y ** 2
    return f * np.exp(-f)
 
n = 10
x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)
z = np.zeros((len(x), len(y)))
for i in range(n):
    for j in range(n):
        z[j, i] = func(x[i], y[j])
print(np.round(z, 2))
pp.figure(figsize = (6, 5))
pp.jet() # a
pp.pcolor(z) # b
pp.colorbar() # c
pp.show()
```
